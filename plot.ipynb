```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
test = pd.read_csv("models/model_v2/test.csv")
```

```python
test
```

```python
# Plot
plt.figure(figsize=(10, 5))
plt.plot(test['episode'], test['reward'])
plt.xlabel("Episode")
plt.ylabel("Reward")
plt.title("Reward vs Episode")
plt.grid(True)
plt.show()
```

```python
# Correlation Coefficients
print(f"Pearson: {test['episode'].corr(test['reward'])}")
print(f"Spearman: {test['episode'].corr(test['reward'], method='spearman')}")
```

```python
# Smoothed plot
test['smooth'] = test['reward'].rolling(window=20).mean()

plt.plot(test['episode'], test['smooth'])
plt.xlabel("Episode")
plt.ylabel("Smoothed Reward")
plt.title("Smoothed Reward Curve")
plt.show()
```
